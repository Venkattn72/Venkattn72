```python
data['Date of Registraion'] = pd.to_datetime(data['DATE_OF_REGISTRATION'])
```

```python

# Extract the year from the date
data['Year'] = data['Date of Registraion'].dt.year

# Count the number of companies registered each year
registration_trends = data.groupby('Year').size()

# Plot the trends
plt.figure(figsize=(12,6))
plt.plot(registration_trends.index, registration_trends.values, marker='o')
plt.xlabel('Year')
plt.ylabel('Number of Companies Registered')
plt.title('Trend of Company Registrations')
plt.grid(True)
plt.show()
```
